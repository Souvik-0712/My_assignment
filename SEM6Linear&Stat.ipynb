6. Use Gauss Elimination to solve the following equations and verify the solution with Numpy’s
built in solver.

    4x+3y-5z=2

    -2x-4y+5z=5

    8x+8y=-3

```python
import numpy as np

# Coefficient Matrix
A = np.array([
    [4, 3, -5],
    [-2, -4, 5],
    [8, 8, 0]
], dtype=float)

# Constant Matrix
B = np.array([2, 5, -3], dtype=float)

# Form the augmented matrix
aug = np.hstack((A, B.reshape(-1, 1)))

print("Augmented Matrix:\n")
print(aug)

n = len(B)

# -----------------------------
# Gaussian Elimination
# -----------------------------
for i in range(n):

    # Make the pivot element 1
    aug[i] = aug[i] / aug[i][i]

    # Eliminate the elements below the pivot
    for j in range(i + 1, n):
        aug[j] = aug[j] - aug[j][i] * aug[i]

print("\nUpper Triangular Matrix:")
print(aug)

# -----------------------------
# Back Substitution
# -----------------------------
x = np.zeros(n)

for i in range(n - 1, -1, -1):
    x[i] = aug[i][-1]
    for j in range(i + 1, n):
        x[i] -= aug[i][j] * x[j]

print("\nSolution using Gaussian Elimination:")
print("x =", x[0])
print("y =", x[1])
print("z =", x[2])

# -----------------------------
# Verification using NumPy
# -----------------------------
solution = np.linalg.solve(A, B)

print("\nSolution using NumPy:")
print(solution)
```

10. Perform Principal Component Analysis (PCA) on the Iris dataset to reduce its dimensionality.
Standardize the features and apply PCA to obtain the first two principal components. Plot the
data using these two components and color the points based on species. Interpret how well the
principal components separate the different species.  
Dataset Link: https://archive.ics.uci.edu/ml/datasets/iris

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Load the Iris dataset
iris = load_iris()

# Create DataFrame
X = pd.DataFrame(iris.data, columns=iris.feature_names)
y = iris.target
species = iris.target_names

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA (reduce to 2 principal components)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Create a DataFrame for visualization
pca_df = pd.DataFrame(X_pca, columns=['Principal Component 1', 'Principal Component 2'])
pca_df['Species'] = [species[i] for i in y]

# Plot the PCA result
plt.figure(figsize=(8,6))

colors = ['red', 'green', 'blue']

for i, target in enumerate(species):
    plt.scatter(
        pca_df[pca_df['Species'] == target]['Principal Component 1'],
        pca_df[pca_df['Species'] == target]['Principal Component 2'],
        color=colors[i],
        label=target,
        s=60
    )

plt.title("PCA of Iris Dataset")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.legend()
plt.grid(True)
plt.show()

# Display explained variance
print("Explained Variance Ratio:", pca.explained_variance_ratio_)
print("Total Variance Explained:", sum(pca.explained_variance_ratio_))
```

16. Using the Iris dataset, perform a one-way ANOVA to test whether there is a significant
difference in sepal length across the three species. State the null and alternative hypotheses,
choose an appropriate significance level, and interpret the results. Also comment on possible
Type I and Type II errors in this context.  
Dataset Link : https://archive.ics.uci.edu/ml/datasets/iris
Hint: Use scipy.stats.f_oneway() on grouped data based on species to perform ANOVA.

```python
# Import libraries
from sklearn.datasets import load_iris
import pandas as pd
from scipy.stats import f_oneway

# Load Iris dataset
iris = load_iris()

# Create DataFrame
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['Species'] = iris.target
df['Species'] = df['Species'].map({
    0: 'Setosa',
    1: 'Versicolor',
    2: 'Virginica'
})

# Group sepal length by species
setosa = df[df['Species'] == 'Setosa']['sepal length (cm)']
versicolor = df[df['Species'] == 'Versicolor']['sepal length (cm)']
virginica = df[df['Species'] == 'Virginica']['sepal length (cm)']

# Perform One-Way ANOVA
F_statistic, p_value = f_oneway(setosa, versicolor, virginica)

print("F-statistic:", F_statistic)
print("P-value:", p_value)

# Decision
alpha = 0.05

if p_value < alpha:
    print("\nReject the Null Hypothesis")
    print("There is a significant difference in mean sepal length among the three species.")
else:
    print("\nFail to Reject the Null Hypothesis")
    print("There is no significant difference in mean sepal length among the three species.")
```

17. A teacher wants to know if three different teaching methods lead to different average exam
scores. She collects data from three groups of students:
Method A: 78, 85, 82, 88
Method B: 72, 75, 68, 70
Method C: 90, 95, 92, 88
Test at the 5% significance level whether the teaching methods produce different mean scores.  
Construct an ANOVA table and make a suitable decision. State the conclusion of the result
obtained.

```python
# Import library
from scipy.stats import f_oneway

# Scores for each teaching method
method_A = [78, 85, 82, 88]
method_B = [72, 75, 68, 70]
method_C = [90, 95, 92, 88]

# Perform One-Way ANOVA
F_statistic, p_value = f_oneway(method_A, method_B, method_C)

# Display results
print("F-statistic:", round(F_statistic, 4))
print("P-value:", p_value)

# Decision at 5% significance level
alpha = 0.05

if p_value < alpha:
    print("\nReject the Null Hypothesis (H0)")
    print("The teaching methods produce significantly different mean scores.")
else:
    print("\nFail to Reject the Null Hypothesis (H0)")
    print("There is no significant difference in the mean scores.")
```

18. A researcher wants to test whether there is an association between gender (Male/Female) and
preference for a product (Like/Dislike). A survey of 100 people yields the following data:
Like
Dislike Total
Male 30
Female 10
Total 40
20
40
60
50
50
100

Test at the 5% significance level whether product preference is independent of gender. State
the conclusion of the result obtained.

```python
# Import libraries
import numpy as np
from scipy.stats import chi2_contingency

# Contingency table
#          Like  Dislike
# Male      30      20
# Female    10      40

table = np.array([
    [30, 20],
    [10, 40]
])

# Perform Chi-Square Test
chi2, p, dof, expected = chi2_contingency(table)

print("Chi-Square Statistic:", round(chi2, 4))
print("Degrees of Freedom:", dof)
print("P-value:", p)

print("\nExpected Frequencies:")
print(expected)

# Decision
alpha = 0.05

if p < alpha:
    print("\nReject the Null Hypothesis (H0)")
    print("There is a significant association between gender and product preference.")
else:
    print("\nFail to Reject the Null Hypothesis (H0)")
    print("Product preference is independent of gender.")
```

20. Using the Iris dataset, draw 100 repeated random samples of size 30 and compute the sample
means of sepal length. Plot the sampling distribution of the sample mean and compare it with
the original distribution. Comment on the behavior of the mean and variance across samples.  
Dataset Link : https://archive.ics.uci.edu/ml/datasets/iris
Hint: Use np.random.choice() to draw samples and plt.hist() to plot the sampling distribution
of means.

```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()

# Create DataFrame
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Select sepal length
sepal_length = df['sepal length (cm)']

# Parameters
sample_size = 30
num_samples = 100

# Draw 100 random samples and compute sample means
sample_means = []

for i in range(num_samples):
    sample = np.random.choice(sepal_length, size=sample_size, replace=True)
    sample_means.append(np.mean(sample))

sample_means = np.array(sample_means)

# Print statistics
print("Population Mean :", round(np.mean(sepal_length), 3))
print("Population Variance :", round(np.var(sepal_length), 3))

print("\nMean of Sample Means :", round(np.mean(sample_means), 3))
print("Variance of Sample Means :", round(np.var(sample_means), 4))

# Plot original distribution
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.hist(sepal_length, bins=10, edgecolor='black')
plt.title("Original Distribution of Sepal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")

# Plot sampling distribution
plt.subplot(1,2,2)
plt.hist(sample_means, bins=10, edgecolor='black')
plt.title("Sampling Distribution of Sample Means")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

21. A company claims that the average lifetime of its LED bulbs is 5,000 hours. A consumer
group suspects the true mean is less than 5,000 hours. They test a random sample of 40 bulbs
and find:
Sample mean (𝑥ˉ) = 4,850 hours
Population standard deviation (𝜎) = 300 hours
Sample size (𝑛) = 40
Significance level (𝛼) = 0.05
Test the hypothesis.
Hint: Calculate the Z-score using $Z = (\bar{x} - \mu) / (\sigma / \sqrt{n})$ and find the p
value using scipy.stats.norm.cdf().

```python
import numpy as np
from scipy.stats import norm

# Given values
x_bar = 4850
mu = 5000
sigma = 300
n = 40
alpha = 0.05

# Calculate Z-score
z = (x_bar - mu) / (sigma / np.sqrt(n))

# Left-tailed p-value
p_value = norm.cdf(z)

print("Z-score:", round(z, 3))
print("P-value:", p_value)

# Decision
if p_value < alpha:
    print("\nReject the Null Hypothesis (H0)")
    print("The average lifetime is significantly less than 5000 hours.")
else:
    print("\nFail to Reject the Null Hypothesis (H0)")
    print("There is insufficient evidence to conclude the average lifetime is less than 5000 hours.")
```

22. A nutritionist claims that a new diet plan reduces average cholesterol levels to below 200
mg/dL. A sample of 25 patients following the diet has:
Sample mean (x ˉ) = 192 mg/dL
Sample standard deviation (s) = 15 mg/dL
Sample size (n) = 25
Significance level (α) = 0.05
Test the claim using a one-sample t-test. State the conclusion of the result obtained.
Hint: Use scipy.stats.ttest_1samp or calculate $t = (\bar{x} - \mu) / (s / \sqrt{n})$ and use
scipy.stats.t.cdf() for the p-value.

```python
import numpy as np
from scipy.stats import t

# Given values
x_bar = 192
mu = 200
s = 15
n = 25
alpha = 0.05

# Calculate t-statistic
t_stat = (x_bar - mu) / (s / np.sqrt(n))

# Degrees of freedom
df = n - 1

# Left-tailed p-value
p_value = t.cdf(t_stat, df)

print("t-statistic:", round(t_stat, 3))
print("Degrees of Freedom:", df)
print("P-value:", round(p_value, 5))

# Decision
if p_value < alpha:
    print("\nReject the Null Hypothesis (H0)")
    print("The average cholesterol level is significantly below 200 mg/dL.")
else:
    print("\nFail to Reject the Null Hypothesis (H0)")
    print("There is insufficient evidence that the average cholesterol level is below 200 mg/dL.")
```

23. Perform a Chi-Square Test of Independence to determine if there is a statistically significant
association between Passenger Class(X)  and Survival (Y) on the Titanic.  Extend the same for
gender and survival.  
Dataset Link : https://www.kaggle.com/c/titanic/data

```python
# Import libraries
import pandas as pd
from scipy.stats import chi2_contingency

# Load Titanic dataset
# Replace 'train.csv' with your file path
df = pd.read_csv("train.csv")

# -----------------------------
# 1. Passenger Class vs Survival
# -----------------------------
table_class = pd.crosstab(df["Pclass"], df["Survived"])

print("Contingency Table: Passenger Class vs Survival")
print(table_class)

chi2, p, dof, expected = chi2_contingency(table_class)

print("\nChi-Square Test (Passenger Class vs Survival)")
print("Chi-square Statistic :", round(chi2, 4))
print("Degrees of Freedom   :", dof)
print("P-value              :", p)

print("\nExpected Frequencies:")
print(pd.DataFrame(expected,
                   index=table_class.index,
                   columns=table_class.columns))

if p < 0.05:
    print("\nReject H0")
    print("Passenger Class and Survival are associated.")
else:
    print("\nFail to Reject H0")
    print("Passenger Class and Survival are independent.")

# -----------------------------
# 2. Gender vs Survival
# -----------------------------
table_gender = pd.crosstab(df["Sex"], df["Survived"])

print("\n\nContingency Table: Gender vs Survival")
print(table_gender)

chi2, p, dof, expected = chi2_contingency(table_gender)

print("\nChi-Square Test (Gender vs Survival)")
print("Chi-square Statistic :", round(chi2, 4))
print("Degrees of Freedom   :", dof)
print("P-value              :", p)

print("\nExpected Frequencies:")
print(pd.DataFrame(expected,
                   index=table_gender.index,
                   columns=table_gender.columns))

if p < 0.05:
    print("\nReject H0")
    print("Gender and Survival are associated.")
else:
    print("\nFail to Reject H0")
    print("Gender and Survival are independent.")
```

24. A fair coin is tossed 10 times. Model the number of heads obtained as a Binomial random
variable. Using Python, compute the probability of getting exactly 6 heads and at most 3 heads.
Plot the probability mass function and interpret the results.
Hint:  use binom from scipy.stats

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters
n = 10
p = 0.5

# (a) Probability of exactly 6 heads
prob_6 = binom.pmf(6, n, p)

# (b) Probability of at most 3 heads
prob_atmost3 = binom.cdf(3, n, p)

print("Probability of exactly 6 heads =", round(prob_6, 5))
print("Probability of at most 3 heads =", round(prob_atmost3, 5))

# Plot Probability Mass Function (PMF)
x = np.arange(0, n + 1)
pmf = binom.pmf(x, n, p)

plt.figure(figsize=(8,5))
plt.bar(x, pmf, color='skyblue', edgecolor='black')
plt.title("Binomial Distribution (n = 10, p = 0.5)")
plt.xlabel("Number of Heads")
plt.ylabel("Probability")
plt.xticks(x)
plt.grid(axis='y')
plt.show()
```
